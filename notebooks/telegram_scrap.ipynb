```python
from telethon import TelegramClient
import logging
# Setup logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Replace these with your own values
api_id = '29256575'
api_hash = '0d56f14d00f288928f28a53baa3e1298'
phone = '+251911834030'

# Use StringSession or modify the session name
client = TelegramClient('new_session_name', '29256575', '0d56f14d00f288928f28a53baa3e1298')

async def scrape_channel(channel_username):
    try:
        await client.start()
        channel = await client.get_entity(channel_username)
        async for message in client.iter_messages(channel):
            # Save the message
            with open('data.csv', 'a', encoding='utf-8') as f:
                f.write(f"{message.date} - {message.sender_id}: {message.text}\n")
            # Download images if present
            if message.photo:
                await message.download_media(file='images/lobelia4cosmetics')
        logger.info(f'Successfully scraped {channel_username}')
    except Exception as e:
        logger.error(f"Error scraping {channel_username}: {e}")

async def main():
    channels = [
        #'DoctorsET',
        #'CheMed123',
        'lobelia4cosmetics',
    ]
    for channel in channels:
        await scrape_channel(channel)

# Use async with to handle the client context
async def run():
    async with client:
        await main()

# Run the event loop
if __name__ == '__main__':
    import asyncio
    #asyncio.run(run())
    await main()








```
